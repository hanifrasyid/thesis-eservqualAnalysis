````python
from google_play_scraper import Sort, reviews
import pandas as pd
from tqdm import tqdm

# Fetch reviews using google_play_scraper, Replace with ur app-id!
app_id = 'com.kai.kaiticketing'

# Fetch reviews
result = []
continuation_token = None
reviews_count = 100000  # change count here

with tqdm(total=reviews_count, position=0, leave=True) as pbar:
    while len(result) < reviews_count:
        apps_reviews, continuation_token = reviews(
            app_id,
            continuation_token=continuation_token,
            lang='id',
            country='id',
            sort=Sort.NEWEST,
            filter_score_with=None,
            count=150
        )
        if not apps_reviews:
            break
        result.extend(apps_reviews)
        pbar.update(len(apps_reviews))
````

````python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

# Opsi 1: panggil melalui file
df_filtering = pd.read_excel("[2] df_caseFolding.xlsx")

# Opsi 2: copy dataframe sebelumnya
# df_filtering = df_caseFolding.copy()

# Mengonversi kolom 'review_translated' menjadi string
df_filtering['review_translated'] = df_filtering['review_translated'].astype(str)

# Fungsi untuk menghapus stop words
def remove_stopwords(rev):
    rev_new = " ".join([i for i in rev if i not in stop_words])
    return rev_new

# Menghapus stop words dari kolom 'review_translated'
df_filtering['review_translated'] = [remove_stopwords(r.split()) for r in df_filtering['review_translated']]

df_filtering.head()
````

#### **4. Cleansing**

````python
import re
import string

# Opsi 1: panggil melalui file
df_cleansing = pd.read_excel("[3] df_filtering.xlsx")

# Opsi 2: copy dataframe sebelumnya
# df_cleansing = df_filtering.copy()

# Fungsi untuk membersihkan teks dari karakter khusus dan non-ASCII
def remove_ulasan_spesial(text):
    # Pengecekan jika text bukan string, konversi ke string
    if not isinstance(text, str):
        text = str(text)
    # Menghapus tab, new line, dan back slice
    text = text.replace('\\t', ' ').replace('\\n', ' ').replace('\\u', ' ').replace('\\', "")
    # Menghapus non ASCII (emoticon, chinese word, etc)
    text = text.encode('ascii', 'replace').decode('ascii')
    # Menghapus mention, link, hashtag
    text = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text).split())
    # Menghapus incomplete URL
    return text.replace("http://", " ").replace("https://", " ")

# Implementasi fungsi cleansing pada DataFrame yang sudah di-filter
df_cleansing['review_translated'] = df_cleansing['review_translated'].astype(str).apply(remove_ulasan_spesial)

df_cleansing.head()
````

````python
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Opsi 1: panggil melalui file
df_lemmatization = pd.read_excel("[7] df_removeShortWord.xlsx")

# Opsi 2: copy dataframe sebelumnya
# df_lemmatization = df_removeShortWord.copy()

# Fungsi untuk mengkonversi tag POS dari format treebank ke format yang diinginkan oleh WordNetLemmatizer
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None  # Hilangkan kata jika tidak sesuai dengan jenis yang diinginkan

# Fungsi untuk melakukan lemmatisasi pada teks
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    words = nltk.word_tokenize(text)  # Tokenisasi teks
    pos_tags = nltk.pos_tag(words)  # POS tagging

    # Lemmatisasi setiap kata dengan tag POS yang sesuai
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in pos_tags if get_wordnet_pos(pos)]  # Lemmatisasi dan filtering
    return ' '.join(lemmatized_words)  # Menggabungkan kata-kata ke dalam satu string

# Pastikan bahwa kolom 'review_translated' adalah string & implementasikan fungsi
df_lemmatization['review_translated'] = df_lemmatization['review_translated'].astype(str).apply(lemmatize_text)
df_lemmatization.head()
````
